```python
import numpy as np
a = np.array([20,30,60,80])
print (np.sum(a))
print (np.cumsum(a))
print (np.product(a))
print (np.cumprod(a))
print (np.min(a))
print (np.max(a))
print (np.mean(a))
print (np.median(a))
```

```python
# For using aggregate functions for 2 variables or say 2 columns present in data.
import numpy as np
price = [100,50,25,200,250]         # list automatically converted to array when using aggregate functions.
quantity = [2,5,8,3,3]

products = np.cumprod([price,quantity], axis = 0)
print (products)
print()
print (products[1])

revenue = np.sum(products[1])
print (f"Total revenue generated by selling all products is: {revenue}")
```
